```python
import pandas as pd

rows = []

for cliente in data['clients']:
    for p in cliente['portfolios']:
        # Valor de portafolio (si falta, suma asset classes/holdings)
        p_val = p.get('value')
        if p_val is None:
            tmp = 0.0
            for ac in p.get('asset_types', []):
                ac_val = ac.get('value')
                if ac_val is None:
                    ac_val = sum((h.get('value') or 0) for h in ac.get('holdings', []))
                tmp += ac_val or 0
            p_val = tmp

        # Peso portafolio normalizado a %
        pw = p.get('weight')
        p_weight = (pw*100 if (pw is not None and abs(pw) <= 1) else (pw or 0))

        for ac in p.get('asset_types', []):
            # Valor asset class (si falta, suma holdings)
            ac_val = ac.get('value')
            if ac_val is None:
                ac_val = sum((h.get('value') or 0) for h in ac.get('holdings', []))

            # Peso asset class normalizado a %
            aw = ac.get('weight')
            ac_weight = (aw*100 if (aw is not None and abs(aw) <= 1) else (aw or 0))

            for h in ac.get('holdings', []):
                hw = h.get('weight')
                rows.append({
                    "Cliente":    cliente.get('name'),
                    "Portafolio": p.get('name'),
                    "Asset_Class": ac.get('type_id'),
                    "Asset":      h.get('asset_id'),

                    # Portafolio
                    "Portfolio_Value":     p_val,
                    "Portfolio_Weight(%)": p_weight,
                    "Portfolio_YTD":       p.get('ytd'),
                    "Portfolio_MTD":       p.get('mtd'),
                    "Portfolio_D1":        p.get('d1'),

                    # Asset class
                    "AC_Value":            ac_val,
                    "AC_Weight(%)":        ac_weight,
                    "AC_YTD":              ac.get('ytd'),
                    "AC_MTD":              ac.get('mtd'),
                    "AC_D1":               ac.get('d1'),

                    # Activo
                    "Value":               h.get('value'),
                    "Weight_Asset(%)":     (hw*100 if (hw is not None and abs(hw) <= 1) else (hw or 0)),
                    "YTD":                 h.get('ytd'),
                    "MTD":                 h.get('mtd'),
                    "D1":                  h.get('d1'),
                })

df = pd.DataFrame(rows)

# Numéricos + NaN -> 0
num_cols = [
    "Portfolio_Value","Portfolio_Weight(%)","Portfolio_YTD","Portfolio_MTD","Portfolio_D1",
    "AC_Value","AC_Weight(%)","AC_YTD","AC_MTD","AC_D1",
    "Value","Weight_Asset(%)","YTD","MTD","D1"
]
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

# Orden final (opcional)
df = df[[
    "Cliente","Portafolio","Asset_Class","Asset",
    "Portfolio_Value","Portfolio_Weight(%)","Portfolio_YTD","Portfolio_MTD","Portfolio_D1",
    "AC_Value","AC_Weight(%)","AC_YTD","AC_MTD","AC_D1",
    "Value","Weight_Asset(%)","YTD","MTD","D1"
]]

df
```

```python
import pandas as pd

# === 1) Selección de portafolios a comparar ===
# Usa tuplas (Cliente, Portafolio). Ejemplos:
targets = [
    ("Inmobiliaria Truro Ltda", "Internacional"),
    ("Modelo2024", "Internacional"),
    # ("Otro Cliente", "Nacional"),
]

# Armo un DataFrame de targets para filtrar rápido
targets_df = pd.DataFrame(targets, columns=["Cliente", "Portafolio"])
df_cmp = df.merge(targets_df, on=["Cliente", "Portafolio"], how="inner")

# === 2) Composición por Asset_Class (valor y % del portafolio) ===
comp = (
    df_cmp.groupby(["Cliente","Portafolio","Asset_Class"], as_index=False)
          .agg(Value=("Value","sum"))
)

# total por (Cliente, Portafolio) para el % dentro del portafolio
tot = comp.groupby(["Cliente","Portafolio"], as_index=False).agg(Total_Portfolio_Value=("Value","sum"))
comp = comp.merge(tot, on=["Cliente","Portafolio"], how="left")
comp["Pct_Portfolio(%)"] = (comp["Value"] / comp["Total_Portfolio_Value"]).fillna(0) * 100

# Tablas “lado a lado” (pivot) por Asset_Class
pivot_value = comp.pivot(index="Asset_Class", columns=["Cliente","Portafolio"], values="Value").fillna(0)
pivot_pct   = comp.pivot(index="Asset_Class", columns=["Cliente","Portafolio"], values="Pct_Portfolio(%)").fillna(0)

# === 3) Métricas a nivel Portafolio (peso y retornos) ===
# Tomamos la primera fila por portafolio porque esas columnas son constantes por portafolio en cada fila
metrics = (
    df_cmp.sort_values(["Cliente","Portafolio"])
          .groupby(["Cliente","Portafolio"], as_index=False)
          .agg(
              Portfolio_Value=("Portfolio_Value","first"),
              Portfolio_Weight_pct=("Portfolio_Weight(%)","first"),
              YTD=("Portfolio_YTD","first"),
              MTD=("Portfolio_MTD","first"),
              D1 =("Portfolio_D1","first"),
          )
)

# === 4) (Opcional) Redondeo para lectura rápida ===
pivot_pct  = pivot_pct.round(2)
metrics[["Portfolio_Weight_pct","YTD","MTD","D1"]] = metrics[["Portfolio_Weight_pct","YTD","MTD","D1"]].round(4)

# === 5) Mostrar resultados ===
print("\n--- Composición por Asset_Class (VALOR) ---")
print(pivot_value)

print("\n--- Composición por Asset_Class (% del portafolio) ---")
print(pivot_pct)

print("\n--- Métricas de Portafolio ---")
print(metrics)
```

```python
import pandas as pd
import unicodedata
import re

# --- 1) Cargar el diccionario desde Excel ---
# cambia la ruta/sheet si corresponde
dic = pd.read_excel("Diccionario Final 2.xlsx")  # o .csv si es CSV

# Nos quedamos con las columnas que necesitamos
cols_keep = ["Ticker", "Geography / Subclass", "Market"]
dic = dic[cols_keep].dropna(subset=["Ticker"]).drop_duplicates("Ticker")

# --- 2) Normalizar strings para evitar fallos por mayúsculas/espacios/acentos ---
def norm(s):
    if pd.isna(s):
        return ""
    s = str(s).strip()
    s = re.sub(r"\s+", " ", s)                  # colapsar espacios
    s = unicodedata.normalize("NFKD", s)        # quitar acentos
    s = "".join(ch for ch in s if not unicodedata.combining(ch))
    return s.upper()

dic["__key__"] = dic["Ticker"].map(norm)

# Tu DF de posiciones (el que armamos antes)
# Si tu nombre de activo está en 'Asset', usamos eso; si está en otra, cámbialo aquí
df["__key__"] = df["Asset"].map(norm)

# --- 3) Merge por clave normalizada ---
df_merged = df.merge(
    dic.rename(columns={"Geography / Subclass": "Geography"})[["__key__", "Geography", "Market"]],
    on="__key__", how="left"
).drop(columns="__key__")

# (Opcional) si quieres ver los que no matchearon:
sin_match = df_merged[df_merged["Geography"].isna()][["Cliente","Portafolio","Asset"]].drop_duplicates()

print(df_merged.head())
```

```python
import pandas as pd

# === 0) Elegir portafolios a comparar ===
# Usa tuplas (Cliente, Portafolio)
targets = [
    ("Inmobiliaria Truro Ltda", "Internacional"),
    ("Modelo2024", "Internacional"),
    # ("Otro Cliente", "Nacional"),
]

targets_df = pd.DataFrame(targets, columns=["Cliente","Portafolio"])
cmp = df_merged.merge(targets_df, on=["Cliente","Portafolio"], how="inner").copy()

# Por si hay NaN numéricos, llévalos a 0 (como preferiste)
num_cols = ["Value","Portfolio_Value","Portfolio_Weight(%)","Portfolio_YTD","Portfolio_MTD","Portfolio_D1",
            "AC_Value","AC_Weight(%)","AC_YTD","AC_MTD","AC_D1","Weight_Asset(%)","YTD","MTD","D1"]
for c in num_cols:
    if c in cmp.columns:
        cmp[c] = pd.to_numeric(cmp[c], errors="coerce").fillna(0)

# === 1) Composición por Asset_Class (VALOR y % del portafolio) ===
# Valor por Asset_Class
ac_val = (cmp.groupby(["Cliente","Portafolio","Asset_Class"], as_index=False)
            .agg(Value_AC=("Value","sum")))

# Total por portafolio para % del portafolio
tot = (ac_val.groupby(["Cliente","Portafolio"], as_index=False)
             .agg(Total_Portfolio_Value=("Value_AC","sum")))

ac = ac_val.merge(tot, on=["Cliente","Portafolio"], how="left")
ac["AC_%_Portfolio"] = (ac["Value_AC"] / ac["Total_Portfolio_Value"]).fillna(0) * 100

# Pivots lado a lado
ac_value_pivot = (ac.pivot(index="Asset_Class", columns=["Cliente","Portafolio"], values="Value_AC")
                    .fillna(0).round(0))
ac_pct_pivot   = (ac.pivot(index="Asset_Class", columns=["Cliente","Portafolio"], values="AC_%_Portfolio")
                    .fillna(0).round(2))

# === 2) Detalle por ACTIVO (peso del activo sobre el PORTAFOLIO) ===
# Calculamos % del activo respecto al portafolio
cmp_tot = (cmp.groupby(["Cliente","Portafolio"], as_index=False)
             .agg(Total_Portfolio_Value=("Value","sum")))
cmp = cmp.merge(cmp_tot, on=["Cliente","Portafolio"], how="left")
cmp["Asset_%_Portfolio"] = (cmp["Value"] / cmp["Total_Portfolio_Value"]).fillna(0) * 100

# Tabla detallada: fila = activo; columnas = portafolio; valores = % del portafolio
# Incluye columnas auxiliares (Asset_Class, Geography, Market)
detail_cols = ["Asset","Asset_Class","Geography","Market","Cliente","Portafolio","Asset_%_Portfolio"]
detail = cmp[detail_cols].drop_duplicates()

# Pivot de % por activo
asset_pct_pivot = (detail.pivot(index=["Asset","Asset_Class","Geography","Market"],
                                columns=["Cliente","Portafolio"],
                                values="Asset_%_Portfolio")
                          .fillna(0).round(2))

# === 3) Métricas de portafolio (peso y retornos) ===
metrics = (cmp.sort_values(["Cliente","Portafolio"])
             .groupby(["Cliente","Portafolio"], as_index=False)
             .agg(Portfolio_Value=("Portfolio_Value","first"),
                  Portfolio_Weight_pct=("Portfolio_Weight(%)","first"),
                  YTD=("Portfolio_YTD","first"),
                  MTD=("Portfolio_MTD","first"),
                  D1 =("Portfolio_D1","first"))
          )
metrics[["Portfolio_Weight_pct","YTD","MTD","D1"]] = metrics[["Portfolio_Weight_pct","YTD","MTD","D1"]].round(4)

# === 4) Mostrar ===
print("\n--- Composición por Asset_Class (VALOR) ---")
print(ac_value_pivot)

print("\n--- Composición por Asset_Class (% del portafolio) ---")
print(ac_pct_pivot)

print("\n--- Detalle por ACTIVO (% del portafolio) ---")
print(asset_pct_pivot)

print("\n--- Métricas de Portafolio ---")
print(metrics)
```
